```python
import pandas as pd
import seaborn as sns
import numpy as np
from pandas import DataFrame
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules

basket = pd.read_csv('/Users/jipyeongseon/Desktop/대학원/산업 빅데이터 분석 실제/python-workspace/Market_Basket_Optimisation.csv', header = None)
basket.head()
```

```python
records = []
for i in range (0, 7501):
    records.append([str(basket.values[i,j]) for j in range(0, 20)])
```

```python
encode = TransactionEncoder()
encoded_array = encode.fit(records).transform(records)

data_frame = pd.DataFrame(encoded_array, columns = encode.columns_)
data_frame
```

```python
basket_clean = data_frame.drop(['nan'], axis=1)
basket_clean
```

```python
frequent_itemsets = apriori(basket_clean, min_support = 0.04, use_colnames = True)
frequent_itemsets.head()
```

```python
pattern_rules = association_rules(frequent_itemsets, metric = 'lift', min_threshold = 1, num_itemsets=1)
pattern_rules
```
