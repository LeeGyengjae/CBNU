```python
import numpy as np

from scipy.stats import pearsonr

alice = np.array([5, 3, 4, 4])
user1 = np.array([3, 1, 2, 3])
user2 = np.array([4, 3, 4, 3])
user3 = np.array([1, 5, 5, 2])

corr = pearsonr(alice, user1)

print(corr)
```

```python
matrix = np.array([[5, 3, 4, 4],
                   [3, 1, 2, 3],
                   [4, 3, 4, 3],
                   [3, 3, 1, 5],
                   [1, 5, 5, 2]
                  ])

corr = np.corrcoef(matrix)

print(corr)
```
